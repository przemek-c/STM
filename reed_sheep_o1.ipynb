```python
# Przykładowy kod obliczający ścieżkę Reed-Shepp dla zadanych (x, y, heading).

import math
import numpy as np
import matplotlib.pyplot as plt

def reed_shepp_path(x, y, heading, turning_radius=1.0, step_size=0.1):
    """
    Funkcja obliczająca prostą ścieżkę Reed-Shepp do punktu (x, y) 
    o zadanym kącie heading dla robota o promieniu skrętu turning_radius.
    Zwraca listę punktów ścieżki.
    """
    # Tu można dodać bardziej rozbudowaną implementację, 
    # np. wykorzystując algorytmy z literatury.
    # Poniżej proste przykładowe generowanie kilku punktów.
    
    num_steps = int(max(abs(x), abs(y)) / step_size)
    path_points = []
    for i in range(num_steps + 1):
        t = i / num_steps
        px = x * t
        py = y * t
        # Tu można uwzględnić heading i manewry cofania
        path_points.append((px, py, heading))

    return path_points

def visualize_path(path_points):
    x_coords = [p[0] for p in path_points]
    y_coords = [p[1] for p in path_points]
    plt.figure()
    plt.plot(x_coords, y_coords, marker='o', label='Reed-Shepp Path')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Reed-Shepp Path Visualization')
    plt.legend()
    plt.axis('equal')
    plt.show()


# Przykładowe wywołanie funkcji:
if __name__ == "__main__":
    x_target = 4.0
    y_target = 1.0
    heading_target = math.radians(180)
    path = reed_shepp_path(x_target, y_target, heading_target)
    print("Obliczona ścieżka Reed-Shepp:", path)
    visualize_path(path)
```
